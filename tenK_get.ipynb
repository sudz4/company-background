```python
import os
from dotenv import load_dotenv
from sec_api import QueryApi

# Load environment variables from .env file
load_dotenv()

# Get SEC_API_KEY environment variable
SEC_API_KEY = os.getenv('SEC_API_KEY')
api = QueryApi(api_key=SEC_API_KEY)

# ticker = "PLTR"

query = {
    "query": { "query_string": { "query": "ticker:PLTR AND formType:\"10-Q\" OR formType:\"10-K\"" } },
    "from": "0",
    "size": "10",
    "sort": [{ "filedAt": { "order": "desc" } }]
}

filings = api.get_filings(query)
```
